# 柱状图

```python
import matplotlib.pyplot as plt
import numpy as np

plt.ylim((-1.1,1))

n=12
X=np.arange(n)
Y1=(1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2=(1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,+Y1)
plt.bar(X,-Y2)

for x,y in zip(X,Y1):
    plt.text(x,y+0.02,'%.2f'%y,ha='center',va='bottom')
    
for x,y in zip(X,Y2):
    plt.text(x,-y-0.02,'-%.2f'%y,ha='center',va='top')

plt.show()
```

## 有序条形图

```python
df_raw = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")
df = df_raw[['cty', 'manufacturer']].groupby('manufacturer').apply(lambda x: x.mean())
```

```python
df
```

```python
df.sort_values('cty',inplace=True)
df.reset_index(inplace=True)
```

```python
import matplotlib.patches as paches

fig,ax=plt.subplots(figsize=(16,10),facecolor='white',dpi=80)
ax.vlines(x=df.index,ymin=0,ymax=df.cty,color='firebrick',alpha=0.7,linewidth=20)


for i,cty in enumerate(df.cty):
    ax.text(i,cty+0.5,round(cty,1),horizontalalignment='center')


ax.set_title('Bar Chart for Highway Mileage', fontdict={'size':22})
ax.set(ylabel='Miles Per Gallon', ylim=(0, 30))
plt.xticks(df.index,df.manufacturer.str.upper(),rotation=60, horizontalalignment='right', fontsize=12)


p1=paches.Rectangle((.57,-0.005),width=.33,height=.13,alpha=.1,facecolor='green',transform=fig.transFigure)
p2=paches.Rectangle((.124,-0.005),width=.446,height=.13,alpha=.1,facecolor='red',transform=fig.transFigure)
fig.add_artist(p1)
fig.add_artist(p2)
plt.show()
```

# 四、分布

## 20 连续变量的直方图

```python
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")
df
```

```python
x_var='displ'
groupby_var='class'
df_agg=df.loc[:,[x_var,groupby_var]].groupby(groupby_var)
vals=[df[x_var].values.tolist()for i,df in df_agg]
vals
```

```python
plt.figure(figsize=(16,9),dpi=80)
colors=[plt.cm.Spectral(i/float(len(vals)-1))for i in range(len(vals))]
n,bins,patches=plt.hist(vals,30,stacked=True,density=False,color=colors[:len(vals)])


plt.legend({group:col for group, col in zip(np.unique(df[groupby_var]).tolist(), colors[:len(vals)])})
plt.title(f"Stacked Histogram of ${x_var}$ colored by ${groupby_var}$", fontsize=22)
plt.xlabel(x_var)
plt.ylabel("Frequency")
plt.ylim(0, 25)
plt.xticks(ticks=bins[::3], labels=[round(b,1) for b in bins[::3]])
plt.show()
```

## 设置图例

```python
plt.plot(x, y, marker="*", linewidth=3, linestyle="--", color="orange")
plt.plot(x, z)
plt.title("matplotlib")
plt.xlabel("height")
plt.ylabel("width")

plt.legend(["Y","Z"],loc="upper right")
```

```python
plt.grid(True)
```

```python
plt.figure(1)
data_clean['薪资'].value_counts().plot(kind='pie')
plt.title('薪资水平统计图')
plt.ylabel('')
plt.rcParams['font.sans-serif']=['Dengxian']
plt.show()


plt.figure(figsize=(15, 6), dpi=80)
x=range(len(data_clean['学历'].value_counts().index))
y1=data_clean['学历'].value_counts().values
y2=data_clean['学历'].value_counts().values
plt.bar(x,y1, width=0.2)
plt.bar([i + 0.2 for i in x],y2, width=0.2)
plt.xticks([i + 0.1 for i in x], data_clean['学历'].value_counts().index)
plt.title('学历与收入的关系统计图')
plt.show()


plt.figure(figsize=(20, 6), dpi=80)
x=range(len(data_clean['职业'].value_counts().index))
y1=data_clean['职业'].value_counts().values
y2=data_clean['职业'].value_counts().values
plt.bar(x,y1, width=0.2)
plt.bar([i + 0.2 for i in x],y2, width=0.2)
plt.xticks([i + 0.1 for i in x], data_clean['职业'].value_counts().index)
plt.title('职业与收入的关系统计图')
plt.show()


plt.figure(figsize=(20,8))
x=data_clean['每周工作时长'].value_counts().index
y=data_clean['资本收益']
plt.scatter(x,y)
plt.xticks(data_clean['每周工作时长'].value_counts().index)
plt.title('所统计人口每小时工作时间的分布图')
plt.show()


plt.figure(figsize=(20,8))
x=range(len(data_clean['每周工作时长'].value_counts().index))
y=data_clean['每周工作时长'].value_counts().values
plt.plot(x,y)
plt.title('每周工作时间与资本收益（capital-gain）之间的关系图')
plt.show()
```

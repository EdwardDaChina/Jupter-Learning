# 折线图

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x=np.arange(9)
y=np.sin(x)
z=np.cos(x)
```

## marker数据点样式，linewidth线宽，linestyle线型样式，color颜色

```python
plt.plot(x,y,linewidth=3,linestyle="--",color="orange")
```

```python
plt.plot(x,z)
```

```python
plt.title("matplotlib")
```

```python
plt.xlabel("height")
plt.ylabel("width")
```

## 设置图例

```python
plt.plot(x, y, marker="*", linewidth=3, linestyle="--", color="orange")
plt.plot(x, z)
plt.title("matplotlib")
plt.xlabel("height")
plt.ylabel("width")

plt.legend(["Y","Z"],loc="upper right")
```

```python
plt.grid(True)
```

```python
plt.plot(x, y, marker="*", linewidth=3, linestyle="--", color="orange")
plt.plot(x, z)
plt.title("matplotlib")
plt.xlabel("height")
plt.ylabel("width")

plt.legend(["Y","Z"],loc="upper right")
plt.grid(True)
plt.show()
```

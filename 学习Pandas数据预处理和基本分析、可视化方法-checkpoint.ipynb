使用drop_duplicates清洗掉重复的数据

```python
data_duplicates=data.drop_duplicates(subset='positionId',keep='first')
data_duplicates
```

drop_duplicates函数通过subset参数选择以哪个列为去重基准。keep参数则是保留方式，first是保留第一个，删除后余重复值，last还是删除前面，保留最后一个

接下来加工salary薪资字段,目的是计算出薪资下限以及薪资上限。

这里需要用到pandas中的apply。它可以针对DataFrame中的一行或者一行数据进行操作，允许使用自定义函数

```python
def cut_word(word):
    position=word.find('-')
    bottomSalary=word[:position-1]
    return bottomSalary
```

```python
data_duplicates.salary.apply(cut_word)
```

我们定义了个word_cut函数，它查找「-」符号所在的位置，并且截取薪资范围开头至K之间的数字，也就是我们想要的薪资上限

「k以上」这类脏数据怎么办呢？find函数会返回-1，如果按照原来的方式截取，是word[:-2]，不是我们想要的结果，所以需要加一个if判断

```python
def cut_word(word):
    position=word.find('-')
    if position != -1:
        bottomSalary=word[:position-1]
    else:
        bottomSalary=word[:word.upper().find('K')]
    return bottomSalary
```

```python
data_duplicates['bottomSalary']=data_duplicates.salary.apply(cut_word)
data_duplicates
```

因为python大小写敏感，我们用upper函数将k都转换为K，然后以K作为截取

```python
data_duplicates.bottomSalary.astype('int')
```

```python
def cut_word(word,method):
    position=word.find('-')
    length=len(word)
    if position != -1:
        bottomSalary=word[:position-1]
        topSalary=word[position+1:length-1]
    else:
        bottomSalary=word[:word.upper().find('K')]
        topSalary=bottomSalary
    if method=='bottom':
        return bottomSalary
    else:
        return topSalary
```

```python
data_duplicates['bottomSalary']=data_duplicates.salary.apply(cut_word,method='bottom')
data_duplicates['topSalary']=data_duplicates.salary.apply(cut_word,method='top')
data_duplicates
```

接下来求平均薪资

```python
data_duplicates.topSalary=data_duplicates.topSalary.astype('int')
data_duplicates.bottomSalary=data_duplicates.bottomSalary.astype('int')
data_duplicates['avgSalary']=data_duplicates.apply(lambda x:(x.bottomSalary+x.topSalary)/2,axis=1)
data_duplicates
```

数据类型转换为数字，这里引入新的知识点，匿名函数lamba; 很多时候我们并不需要复杂地使用def定义函数，而用lamdba作为一次性函数

前面的lambda x:理解为输入，后面的星号区域则是针对输入的x进行运算。案例中，因为同时对top和bottom求平均值，所以需要加上x.bottomSalary和x.topSalary。word_cut的apply是针对Series，现在则是DataFrame

这里的lambda可以用(df_duplicates.bottomSalary + df_duplicates.topSalary)/2替代

但是在pandas的分组计算中，它也不是最灵活的函数。

```python
def topN(data,n=5):
    counts=data.value_counts()
    return counts.sort_values(ascending=False)[:n]
```

```python
data_clean.groupby('city').companyShortName.apply(topN)
```

另外一种分析思路是对数据进行深加工

我们将薪资设立出不同的level

```python
bins=[0,3,5,10,15,20,30,100]
level=['0-3','3-5','5-10','10-15','15-20','20-30','30-100']
data_clean['level']=pd.cut(data_clean['avgSalary'],bins=bins,labels=level)
```

```python
data_clean[['avgSalary','level']]
```

cut的作用是分桶，它也是数据分析常用的一种方法，将不同数据划分出不同等级，也就是将数值型数据加工成分类数据，在机器学习的特征工程中应用比较多

cut可以等距划分，传入一个数字就好,这里为了更好的区分，传入了一组列表进行人工划分，加工成相应的标签

```python
data_level=data_clean.groupby(['city','level']).avgSalary.count().unstack()
data_level_prop=data_level.apply(lambda x:x/x.sum(),axis=1)
data_level_prop.plot.bar(stacked=True,figsize=(15,10))
```
